```python
import pandas as pd
import numpy as np
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
 
g.add_edge(0, 2)
g.add_edge(2, 1)
g.add_edge(1, 7)
g.add_edge(7, 6)
g.add_edge(6, 0)

g.add_edge(0, 4)
g.add_edge(4, 5)
g.add_edge(5, 9)
g.add_edge(9, 10)
g.add_edge(10, 0)

g.add_edge(0,3)
g.add_edge(3,8)
g.add_edge(8,0)

labels = {0:'Depot = 0',1:'P2 = 30',2:'P2 = 60',3:'P2 = 50',4:'P1 = 60',5:'P1 = 50',6:'C1 = -30',7:'C1 = -60',8:'C1 = -50',9:'C2 = -60',10:'C2 = -50'}   

pos=nx.spring_layout(g)
plt.figure(figsize=(12,12))
plt.title("Standard Last Mile Delivery")
nx.draw_networkx(g,pos,with_labels=True,node_size=1000,node_color='#000000',font_color='#ffffff',arrows=True) 
for key,val in pos.items():
  val[1]+=0.08

nx.draw_networkx_labels(g,pos,labels,font_size=16,font_color='#000000')

plt.savefig("standard.png")
```
